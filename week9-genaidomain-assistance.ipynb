```python
import os
from openai import OpenAI
from kaggle_secrets import UserSecretsClient

# Fetch the secret API key safely from Kaggle's storage
user_secrets = UserSecretsClient()
api_key = user_secrets.get_secret("OPENAI_API_KEY")

# Initialize the OpenAI client
client = OpenAI(api_key=api_key)
print('OpenAI client successfully initialized!')
```

**Part 1.2 — Your First API Call**

```python
# Task 1.2: Make First API Call using gpt-4
try:
    response = client.chat.completions.create(
        model='gpt-4',
        messages=[
            {
                'role': 'user', 
                'content': 'Say hello and introduce yourself as an AI assistant!'
            }
        ],
        max_tokens=150,
        temperature=0.7
    )

    # Extract and print response message
    message = response.choices[0].message.content
    print(message)
except Exception as e:
    print(f"An error occurred: {e}")
try:
    response = client.chat.completions.create(
        model='gpt-4',
        messages=[
            {
                'role': 'user', 
                'content': 'Say hello and introduce yourself as an AI assistant!'
            }
        ],
        max_tokens=150,
        temperature=0.7
    )

    # Extract and print response message
    message = response.choices[0].message.content
    print(message)
except Exception as e:
    print(f"An error occurred: {e}")
```

**Part 2 — Core Chat Function & System Prompt**

```python
def get_gpt_response(messages, model='gpt-3.5-turbo'):
    """Sends a list of message history to GPT and returns the text response."""
    try:
        response = client.chat.completions.create(
            model=model, # gpt-3.5-turbo is faster and cheaper for iteration loops
            messages=messages,
            max_tokens=500,
            temperature=0.7
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error connecting to OpenAI: {e}"

def start_chat_session(system_prompt):
    """
    Starts an interactive conversation loop in the notebook.
    Type 'quit' or 'exit' to end the conversation.
    """
    # Initialize conversation history with the system role configuration
    messages = [{'role': 'system', 'content': system_prompt}]
    
    print('Chatbot ready! Type "quit" or "exit" to end the conversation.\n')
    
    while True:
        user_input = input('You: ')
        if user_input.lower() in ['quit', 'exit']:
            print('Goodbye!')
            break
            
        if not user_input.strip():
            continue
            
        # Append user message to history
        messages.append({'role': 'user', 'content': user_input})
        
        # Get AI response
        assistant_response = get_gpt_response(messages)
        
        # Append assistant response to history to maintain multi-turn dialogue
        messages.append({'role': 'assistant', 'content': assistant_response})
        
        print(f'Assistant: {assistant_response}\n')
```

**Part 3.1 — Create & Run the HR Assistant**

```python
# Task 3.1: Define the HR System Prompt and policies
hr_system_prompt = """You are an HR assistant for a technology company.
Company policies:
Vacation: 15 days per year
Sick leave: Unlimited (with manager approval)
Remote work: 3 days per week
Health insurance: Fully covered
401(k) matching: Up to 5%

Your role:
1. Answer employee questions about policies
2. Be friendly and supportive
3. If unsure, suggest contacting HR department
4. Keep responses concise (2-3 sentences)"""

# Start the interactive HR chat loop
start_chat_session(hr_system_prompt)
```

**Part 3.2 — Create & Run the Customer Support Bot**

```python
# Task 3.2: Define Customer Support System Prompt and policies
support_system_prompt = """You are a customer support agent for TechShop, an electronics retailer.
Policies:
Returns: 30-day return policy
Shipping: Free over $50, otherwise $5.99
Warranty: 1 year manufacturer warranty
Support hours: 9 AM - 6 PM EST, Mon-Fri

Your tone: 
Empathetic and patient
Apologize when appropriate
Offer to escalate complex issues
Solution-focused"""

# Start the interactive Customer Support chat loop
start_chat_session(support_system_prompt)
```
